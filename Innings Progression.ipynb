```python
import pandas as p
```

```python
from matplotlib import pyplot as plt
```

```python
sample_data=p.read_csv('Innings progression.csv')
```

```python
 sample_data
    
```

```python
Ind=sample_data[sample_data.Team=='India']
Ind
```

```python
Pak=sample_data[sample_data.Team=='Pakistan']
Pak
```

```python
plt.plot(Ind.Overs,Ind.Runs)
plt.show()
```

```python
plt.plot(Pak.Overs,Pak.Runs)
plt.show()
```

```python
plt.title('Innings Progression')
plt.plot(Ind.Overs,Ind.Runs)
plt.plot(Pak.Overs,Pak.Runs)
plt.legend(['India','Pakistan'])
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.show()
```
